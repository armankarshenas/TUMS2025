# Introduction to NumPy, and Matplotlib 

## **What are Libraries in Python?**
Libraries in Python are collections of pre-written code that provide functionality for performing specific tasks, such as numerical computation, data visualization, or web development. By using libraries, you can avoid writing code from scratch and save time while benefiting from optimized, well-tested solutions.

Libraries are usually imported at the beginning of your Python script or notebook using the `import` statement. For example:
```python
import numpy as np
import matplotlib.pyplot as plt
```
Here, `numpy` is imported with the alias `np`, and `matplotlib.pyplot` is imported as `plt` for convenience and readability.

---

## **NumPy**: 
NumPy (short for Numerical Python) is a powerful library for numerical computing in Python. It provides support for working with arrays, matrices, and high-level mathematical functions to perform operations on these data structures. NumPy is essential for scientific computing, enabling efficient manipulation of large datasets.

To use NumPy in your Python programs, it must first be imported. By convention, it is typically imported as `np` for brevity:
```python
import numpy as np
```

---

## **Matplotlib**:
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It allows you to create a wide variety of plots such as line plots, bar charts, histograms, scatter plots, and more. A common submodule used for plotting is `pyplot`, which is conventionally imported as `plt`:
```python
import matplotlib.pyplot as plt
```

With Matplotlib, you can visualize your data and make your analyses more understandable.

---

## **User-Defined Functions**:
In Python, you can define your own functions to modularize and reuse code. A user-defined function is written using the `def` keyword and can be stored in a Python file (with a `.py` extension). This makes it reusable across different scripts or projects.

Here’s an example of a Python file (`my_functions.py`) with a user-defined function:
```python
# my_functions.py
def greet(name):
    return f"Hello, {name}!"
```

You can then import and use this function in your program:
```python
from my_functions import greet

message = greet("Class")
print(message)  # Output: Hello, Class!
```

---

### Key Points:
1. Libraries like NumPy and Matplotlib must be imported to use their functionality.
2. User-defined functions help you create reusable, modular code.
3. Use import statements to bring external libraries and functions from other files into your script.


# Understanding NumPy: Arrays, Operations, and Linear Algebra

## **Introduction to NumPy Arrays**
NumPy arrays are the core feature of the library, providing a fast and flexible way to store and manipulate numerical data. Unlike Python lists, NumPy arrays are homogenous (all elements are of the same type) and support element-wise operations, which makes them highly efficient for numerical computations.

### Creating NumPy Arrays
You can create NumPy arrays using the `np.array()` function or specialized functions like `np.zeros()`, `np.ones()`, or `np.arange()`:
```python
import numpy as np

# Creating an array from a list
array = np.array([1, 2, 3, 4, 5])

# Creating a 2D array (matrix)
matrix = np.array([[1, 2], [3, 4]])

# Creating arrays with zeros, ones, or a range of numbers
zeros = np.zeros((2, 3))
ones = np.ones((3, 2))
range_array = np.arange(0, 10, 2)  # [0, 2, 4, 6, 8]
```

---

## **Important Array Operations**
NumPy provides a wide variety of operations for element-wise and matrix-level manipulations.

### Element-wise Operations
```python
array = np.array([1, 2, 3, 4])

# Element-wise operations
squared = array ** 2  # [1, 4, 9, 16]
sum_array = array + 10  # [11, 12, 13, 14]
product_array = array * 2  # [2, 4, 6, 8]
```

### Aggregation Operations
```python
array = np.array([1, 2, 3, 4])

# Sum, mean, and maximum
sum_val = np.sum(array)  # 10
mean_val = np.mean(array)  # 2.5
max_val = np.max(array)  # 4
```

---

## **Matrix Multiplications and Linear Algebra**
NumPy supports advanced matrix operations and linear algebra functions that are essential for scientific computing.

### Matrix Multiplications
Matrix multiplication is performed using the `@` operator or `np.dot()`:
```python
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])

# Matrix multiplication
result = matrix1 @ matrix2  # or np.dot(matrix1, matrix2)
```

### Transpose and Inverse
You can transpose a matrix or compute its inverse:
```python
matrix = np.array([[1, 2], [3, 4]])

# Transpose
transpose = matrix.T

# Inverse
inverse = np.linalg.inv(matrix)
```

### Eigenvalues and Eigenvectors
NumPy provides functions for more advanced linear algebra operations:
```python
matrix = np.array([[1, 2], [2, 3]])

eigenvalues, eigenvectors = np.linalg.eig(matrix)
```

### Solving Linear Systems
Solve systems of linear equations using `np.linalg.solve()`:
```python
A = np.array([[3, 1], [1, 2]])
B = np.array([9, 8])

# Solving Ax = B
x = np.linalg.solve(A, B)  # x contains the solution
```

---

## **Key Takeaways**
- NumPy arrays are efficient and powerful for numerical computations.
- They support element-wise operations and matrix-level manipulations.
- NumPy’s linear algebra module (`np.linalg`) provides tools for solving equations, finding eigenvalues, and more.

# Introduction to Matplotlib: Plotting Basics

Matplotlib is a versatile library for creating visualizations in Python. It allows users to generate static, animated, and interactive plots to visualize data effectively. This section introduces the basics of Matplotlib and demonstrates how to create commonly used plots.

## **Getting Started with Matplotlib**
To use Matplotlib, you need to import the `pyplot` submodule. By convention, it is imported as `plt`:
```python
import matplotlib.pyplot as plt
```

### Anatomy of a Matplotlib Plot
A typical Matplotlib plot consists of the following elements:
- **Figure**: The overall container for the plot.
- **Axes**: The area where data is plotted (can have multiple axes in a figure).
- **Title**: Describes the overall purpose of the plot.
- **Labels**: Axis labels (e.g., x-axis and y-axis).
- **Legend**: Explains what the plotted data represents.

---

## **Basic Plotting**

### Line Plot
Line plots are used to show trends over a continuous variable, such as time.
```python
# Example: Simple Line Plot
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

plt.plot(x, y, label="y = x^2")
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

### Scatter Plot
Scatter plots are used to display individual data points and their relationships.
```python
# Example: Scatter Plot
x = [1, 2, 3, 4, 5]
y = [5, 7, 6, 8, 7]

plt.scatter(x, y, color='red', label="Data Points")
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

### Bar Plot
Bar plots are useful for comparing categories.
```python
# Example: Bar Plot
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

plt.bar(categories, values, color='blue')
plt.title("Bar Plot Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

### Histogram
Histograms show the distribution of a dataset.
```python
# Example: Histogram
import numpy as np

data = np.random.randn(1000)  # Generate random data

plt.hist(data, bins=20, color='green', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.show()
```

### Multiple Plots
You can combine multiple plots in a single figure using subplots.
```python
# Example: Subplots
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [25, 16, 9, 4, 1]

plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
plt.plot(x, y1, label="y = x^2")
plt.title("Plot 1")
plt.legend()

plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
plt.plot(x, y2, label="y = 26-x^2", color='orange')
plt.title("Plot 2")
plt.legend()

plt.tight_layout()
plt.show()
```

---

## **Customizing Plots**
Matplotlib allows extensive customization of plots:
- **Colors**: Change line or bar colors with keywords like `color='red'`.
- **Markers**: Use markers to highlight data points, e.g., `marker='o'`.
- **Line Styles**: Use `linestyle='--'` for dashed lines, `'-'` for solid lines, etc.
- **Grid**: Add a grid using `plt.grid(True)`.

Example:
```python
# Customized Plot
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

plt.plot(x, y, color='purple', linestyle='--', marker='o', label="y = x^2")
plt.title("Customized Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.legend()
plt.show()
```

---

## **Key Takeaways**
1. Matplotlib is a versatile library for creating visualizations in Python.
2. It provides various plot types such as line plots, scatter plots, bar plots, and histograms.
3. Customization options allow you to tailor your plots to meet specific needs.
4. Subplots enable combining multiple visualizations in one figure.

# Example 3: Simulating Bacterial Growth Using the Euler Method

Bacterial growth can often be modeled mathematically using differential equations. One common model is the exponential growth equation:

$$ \frac{dN}{dt} = r \cdot N $$

Where:
- $ N $ is the population size at time $ t $.
- $ r $ is the growth rate.
- $ \frac{dN}{dt} $ is the rate of change of the population over time.

In this example, we’ll simulate bacterial growth numerically using the **Euler method**, a simple and effective way to approximate solutions to differential equations.

---

## **The Euler Method**
The Euler method approximates the solution to a differential equation by breaking it into small time steps $ \Delta t $. The population at the next time step $ N(t + \Delta t) $ is estimated using:

$$ N(t + \Delta t) = N(t) + \Delta t \cdot \frac{dN}{dt} $$

This iterative process allows us to calculate $ N $ at successive time points.

---

## **Simulating Bacterial Growth**

### Parameters
We’ll define the following parameters for the simulation:
- Initial population size $ N_0 $.
- Growth rate $ r $.
- Total simulation time $ T $.
- Time step $ \Delta t $.

---

## **Experimenting with Parameters**
- **Initial Population**: Try different values of $ N_0 $ to see how starting conditions influence growth.
- **Growth Rate**: Adjust $ r $ to simulate slower or faster growth.
- **Time Step**: Use smaller $ \Delta t $ for more accurate results, but note that smaller steps increase computational time.
